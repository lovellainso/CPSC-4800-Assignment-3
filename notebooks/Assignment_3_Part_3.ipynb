```python
# 3 Relationship Analysis
# Lets take a look on the relationship between the sale price and the year sold
data[["SalePrice", "YrSold"]].describe()
```

```python
data[["SalePrice", "YrSold"]].groupby("YrSold").mean()
# mean value grouped by year sold
```

```python
 np.corrcoef(data['SalePrice'], data['YrSold'])[0,1]   #coefficient = -0.0289
# We can see that the correlation coefficient between these two variables is -0.0289, which means that there is a weak
# relationship between these two variables. Thus, they are somehow not related to each other, any increase or decrease in price
# is not associated with the year sold.
```

```python
sns.scatterplot(x="YrSold", y="SalePrice", data=data);
```

```python
sns.lmplot(x="YrSold", y="SalePrice", data=data);
```
